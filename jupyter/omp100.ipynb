Średnia wartość czasu obliczeń

```python
mean_duration = []
for i in range(1, 9):
    mean_duration.append(data.loc[data.nr_of_threads == i].duration.mean())
```

```python
plt.plot(range(1, 9), mean_duration, 'bo')
plt.title('Średnia wartość czasu trwania programu dla różnej ilości wątków 100 węzłów')
plt.ylabel('Mean duration [s]')
plt.xlabel('Number of threads')
```

Współczynnik przyśpieszenia 
S(n, p) = T(n, 1)/T(n, p)

```python
speed_up = [mean_duration[0]/i for i in mean_duration]
```

```python
speed_up
```

Wydajnosc S(n, p) = S(n, p)/p *100%

```python
wydajnosc = [i/(idx+1)*100 for idx, i in enumerate(speed_up)]
```

```python
wydajnosc
```

```python
plt.plot(range(1, 9), speed_up,  label = "wsp. przyśpieszenia")
plt.title('Wskaźniki jakości')
plt.xlabel('Number of threads')
plt.legend()
```

```python
plt.plot(range(1, 9), wydajnosc,  label = "wydajność")
plt.title('Wskaźniki jakości')
plt.xlabel('Number of threads')
plt.legend()
```
